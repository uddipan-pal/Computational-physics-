```python
import numpy as np
#set 1:
A1 = np.array([[3, -1, 1],
               [3, 6, 2],
               [3, 3, 7]])
b1 = np.array([1, 0, 4])

# Set 2: Coefficient matrix A and constant vector b
A2 = np.array([[10, -1, 0],
               [-1, 10, -2],
               [0, -2, 10]])
b2 = np.array([9, 7, 6])

# Set 3: Coefficient matrix A and constant vector b
A3 = np.array([[10, 5, 0, 0],
               [5, 10, -4, 0],
               [0, -4, 8, -1],
               [0, 0, -1, 5]])
b3 = np.array([6, 25, -11, -11])

# Set 4: Coefficient matrix A and constant vector b
A4 = np.array([[4, 1, 1, 0, 1],
               [-1, -3, 1, 1, 0],
               [2, 1, 5, -1, -1],
               [-1, -1, -1, 4, 0],
               [0, 2, -1, 1, 4]])
b4 = np.array([6, 6, 6, 6, 6])
```

```python
#q1
import numpy as np
def pri(*args):
    for arg in args:
        print(arg)

# Example function to be decorated


def relax(A,b,w): #relaxation method
  n=len(A)
  x=np.zeros(n)
  print("x0=",x)
  print("r0=",b)
  for k in range(2):
    x_old=x.copy()
    print("\niteration: ",k+1)
    for i in range(n):
      s1=0.0
      for j in range(n):
        s1=s1+A[i][j]*x[j] #kth iteration value

      r=b[i]-s1 #residual vector component
      x[i]=x_old[i]+(w*r)/A[i][i]
      print("r_",i+1,"=",r)
      print("x_",i+1,"=",x[i])
  print("\nsolution:x=")
  return x
w=1.3
print("Question 5")
print("relaxation method(w=",w,"):\n")
pri("A=",A1)
print("b=",b1)
print(relax(A1,b1,w) )
print("---------------")
pri("A=",A2)
print("b=",b2)
print(relax(A2,b2,w) )
print("---------------")
pri("A=",A3)
print("b=",b3)
print(relax(A3,b3,w) )
print("---------------")
pri("A=",A4)
print("b=",b4)
print(relax(A4,b4,w) )
print("---------------")
```
